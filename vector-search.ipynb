```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd

books = pd.read_csv("books_cleaned.csv")
```

```python
books
```

```python
books["tagged_description"].to_csv("tagged_description.txt",
                                   index=False,
                                   header=False,
                                   lineterminator="\n"
)
```

```python
raw_documents = TextLoader("tagged_description.txt", encoding="utf-8").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
db_books = Chroma.from_documents(
    documents,
    embedding=HuggingFaceEmbeddings()
)
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip().replace('"', ''))]
```

```python
def retrive_semantic_recommendations(
        query: str,
        db_connector: Chroma,
        top_k: int = 10
) -> pd.DataFrame:
    recs = db_connector.similarity_search(query, k=50)

    books_list = []
    for i in range(0, len(recs)):

        isbn_str = recs[i].page_content.split()[0].strip().replace('"', '')
        try:
            books_list.append(int(isbn_str))
        except ValueError:
            print(f"Warning: Could not convert '{isbn_str}' to int. Skipping.")
            continue

    return books[books["isbn13"].isin(books_list)].head(top_k)
```

```python
retrive_semantic_recommendations("A book to teach children about nature",  db_connector=db_books)
```
